## 3) Ð—Ð°Ð³Ñ€ÑƒÐ·ÐºÐ° ÐºÑ€Ð¸Ð¿Ñ‚Ð¾Ð²Ð°Ð»ÑŽÑ‚ Ñ‡ÐµÑ€ÐµÐ· CoinGecko API
ÐœÑ‹ Ð·Ð°Ð¿Ñ€Ð¾ÑÐ¸Ð¼ Ð¸ÑÑ‚Ð¾Ñ€Ð¸Ñ‡ÐµÑÐºÐ¸Ðµ Ñ†ÐµÐ½Ñ‹ Ð² Ð´Ð¾Ð»Ð»Ð°Ñ€Ð°Ñ…, Ð·Ð°Ñ‚ÐµÐ¼ Ð¿ÐµÑ€ÐµÐ²ÐµÐ´Ñ‘Ð¼ Ð² Ñ€ÑƒÐ±Ð»Ð¸ Ð¿Ð¾ ÐºÑƒÑ€ÑÑƒ USD/RUB.
Ð•ÑÐ»Ð¸ Ñ…Ð¾Ñ‡ÐµÑˆÑŒ Ð¸ÑÐ¿Ð¾Ð»ÑŒÐ·Ð¾Ð²Ð°Ñ‚ÑŒ Ñ€ÑƒÐ±Ð»Ð¸ Ð¿Ñ€ÑÐ¼Ð¾, Ð¼Ð¾Ð¶Ð½Ð¾ Ð¿Ð¾Ð»ÑƒÑ‡Ð¸Ñ‚ÑŒ Ñ€Ñ‹Ð½Ð¾Ðº / tickers Ñ MOEX (ÑÐ¿Ð±) Ð¸Ð»Ð¸ ÐºÐ¾Ð½Ð²ÐµÑ€Ñ‚Ð°Ñ†Ð¸ÑŽ.

```python
# !pip install pycoingecko
```

```python
# Ð˜Ð¼Ð¿Ð¾Ñ€Ñ‚ Ð±Ð¸Ð±Ð»Ð¸Ð¾Ñ‚ÐµÐº
from pycoingecko import CoinGeckoAPI
import pandas as pd
import yfinance as yf
import time

cg = CoinGeckoAPI()

# Ð¡Ð¿Ð¸ÑÐ¾Ðº ÐºÑ€Ð¸Ð¿Ñ‚Ð¾Ð²Ð°Ð»ÑŽÑ‚ Ð´Ð»Ñ Ð°Ð½Ð°Ð»Ð¸Ð·Ð° (toncoin Ð½ÐµÑ‚ Ð² Ð² CoinGeckoAPI)
crypto_ids = ['bitcoin', 'ethereum'] #'toncoin']  # Ð¼Ð¾Ð¶Ð½Ð¾ Ð´Ð¾Ð±Ð°Ð²Ð¸Ñ‚ÑŒ Ð´Ñ€ÑƒÐ³Ð¸Ðµ Ð¿Ð¾ Ð¶ÐµÐ»Ð°Ð½Ð¸ÑŽ
vs_currency = 'usd'

# Ð¤ÑƒÐ½ÐºÑ†Ð¸Ñ Ð´Ð»Ñ Ð·Ð°Ð³Ñ€ÑƒÐ·ÐºÐ¸ Ð¸ÑÑ‚Ð¾Ñ€Ð¸Ñ‡ÐµÑÐºÐ¸Ñ… Ð´Ð°Ð½Ð½Ñ‹Ñ…
def get_crypto_history(crypto_id):
    print(f'Ð—Ð°Ð³Ñ€ÑƒÐ¶Ð°ÐµÐ¼ Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð´Ð»Ñ {crypto_id}...')
    data = cg.get_coin_market_chart_by_id(id=crypto_id, vs_currency=vs_currency, days='365')
    df = pd.DataFrame(data['prices'], columns=['timestamp', crypto_id])
    df['date'] = pd.to_datetime(df['timestamp'], unit='ms')
    df.set_index('date', inplace=True)
    df.drop('timestamp', axis=1, inplace=True)
    df = df.resample('D').last().ffill()  # Ð´Ð½ÐµÐ²Ð½Ð¾Ð¹ Ð¸Ð½Ñ‚ÐµÑ€Ð²Ð°Ð», Ð·Ð°Ð¿Ð¾Ð»Ð½ÑÐµÐ¼ Ð¿Ñ€Ð¾Ð¿ÑƒÑÐºÐ¸
    time.sleep(1)  # Ñ‡Ñ‚Ð¾Ð±Ñ‹ Ð½Ðµ Ð¿ÐµÑ€ÐµÐ³Ñ€ÑƒÐ¶Ð°Ñ‚ÑŒ API
    return df

# Ð—Ð°Ð³Ñ€ÑƒÐ¶Ð°ÐµÐ¼ Ð²ÑÐµ ÐºÑ€Ð¸Ð¿Ñ‚Ð¾Ð²Ð°Ð»ÑŽÑ‚Ñ‹ Ð¸ Ð¾Ð±ÑŠÐµÐ´Ð¸Ð½ÑÐµÐ¼ Ð² Ð¾Ð´Ð¸Ð½ DataFrame
crypto_list = []
for cid in crypto_ids:
    crypto_list.append(get_crypto_history(cid))

crypto_df = pd.concat(crypto_list, axis=1)
print(f"Ð”Ð°Ð½Ð½Ñ‹Ðµ Ð¿Ð¾ ÐºÑ€Ð¸Ð¿Ñ‚Ð¾Ð²Ð°Ð»ÑŽÑ‚Ð°Ð¼ Ð·Ð°Ð³Ñ€ÑƒÐ¶ÐµÐ½Ñ‹: Ð Ð°Ð·Ð¼ÐµÑ€ - {crypto_df.shape}")
crypto_df.head()
```

```python
# toncoin Ð²Ð¾Ð·ÑŒÐ¼ÐµÐ¼ Ð¸Ð· YFinance
ton_df = yf.download('TON-USD', period='1y', interval='1d')['Close']
ton_df.rename(columns={'TON-USD': 'toncoin'}, inplace=True)

# ÐŸÑ€ÐµÐ¾Ð±Ñ€Ð°Ð·ÑƒÐµÐ¼ Ð² DataFrame
ton_df.index = pd.to_datetime(ton_df.index)
ton_df.head()
```

```python
# ÐžÐ±ÑŠÐµÐ´Ð¸Ð½Ð¸Ð¼ Ð·Ð°Ð¿Ð¸ÑÐ¸ ÐºÑ€Ð¸Ð¿Ñ‚Ð¾Ð²Ð°Ð»ÑŽÑ‚ Ð² ÐµÐ´Ð¸Ð½Ñ‹Ð¹ Ð´Ð°Ñ‚Ð°Ñ„Ñ€ÐµÐ¹Ð¼
crypto_df = pd.concat([crypto_df, ton_df], axis=1).dropna()
crypto_df
```

## 8) Ð¤Ð¾Ñ€Ð¼Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð¸Ðµ Ð¿Ð¾Ñ€Ñ‚Ñ„ÐµÐ»Ñ Ð¸ Ð¾Ð¿Ñ‚Ð¸Ð¼Ð¸Ð·Ð°Ñ†Ð¸Ñ (Ð¿Ñ€Ð¾ÑÑ‚Ð°Ñ MV Ð¾Ð¿Ñ‚Ð¸Ð¼Ð¸Ð·Ð°Ñ†Ð¸Ñ)
ÐœÐ¾Ð¶Ð½Ð¾ Ñ€ÐµÐ°Ð»Ð¸Ð·Ð¾Ð²Ð°Ñ‚ÑŒ Ð½Ð° Ð¾ÑÐ½Ð¾Ð²Ðµ ÑÑ€ÐµÐ´Ð½Ð¸Ñ… Ð´Ð¾Ñ…Ð¾Ð´Ð½Ð¾ÑÑ‚ÐµÐ¹ Ð¸ ÐºÐ¾Ð²Ð°Ñ€Ð¸Ð°Ñ†Ð¸Ð¸ (Markowitz). ÐŸÑ€Ð¸Ð¼ÐµÑ€ Ð½Ð¸Ð¶Ðµ.

```python
# ÐŸÑ€Ð¾ÑÑ‚Ð°Ñ MV-Ð¾Ð¿Ñ‚Ð¸Ð¼Ð¸Ð·Ð°Ñ†Ð¸Ñ (Ñ€Ð°Ð½Ð´Ð¾Ð¼Ð½Ñ‹Ðµ Ð¿Ð¾Ñ€Ñ‚Ñ„ÐµÐ»Ð¸)
import numpy as np

def random_portfolios(returns, n_portfolios=5000, risk_free=0.0):
    mu = returns.mean() * 252
    cov = returns.cov() * 252
    results = []
    for _ in range(n_portfolios):
        w = np.random.random(len(mu))
        w /= w.sum()
        ret = np.dot(w, mu)
        vol = np.sqrt(np.dot(w.T, np.dot(cov, w)))
        sharpe = (ret - risk_free) / vol if vol != 0 else 0
        results.append({'w': w, 'ret': ret, 'vol': vol, 'sharpe': sharpe})
    return results

# usage (Ð¿Ð¾ÑÐ»Ðµ Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÐµÐ½Ð¸Ñ returns):
# results = random_portfolios(returns)
# best = max(results, key=lambda x: x['sharpe'])
# print('Best Sharpe', best['sharpe'])
```
